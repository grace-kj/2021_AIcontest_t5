```python
##############개체명 인식

import pandas as pd
q_dataset= pd.read_csv("/content/drive/MyDrive/Colab Notebooks/국립국어원/ner_result_q.tsv", delimiter='\t')
q_dataset.head()
text_dataset= pd.read_csv("/content/drive/MyDrive/Colab Notebooks/국립국어원/ner_result_text.tsv", delimiter='\t')
text_dataset.head()
```

```python
train_data= pd.read_csv("/content/drive/MyDrive/Colab Notebooks/국립국어원/SKT_BoolQ_Train.tsv", delimiter='\t')
dev_data= pd.read_csv("/content/drive/MyDrive/Colab Notebooks/국립국어원/SKT_BoolQ_Dev.tsv", delimiter='\t')
```

```python
train_data.head()
```

```python
#text_dataset에서 train, dev 구분

is_train=[]
is_dev=[]

for item in (text_dataset['ID']):
  if item[:5]=='train':
    is_train.append(True)
    is_dev.append(False)
  else:
    is_train.append(False)
    is_dev.append(True)

train_text= text_dataset[is_train]
dev_text= text_dataset[is_dev]
```

```python
#q_dataset에서 train, dev 구분

is_train=[]
is_dev=[]

for item in (q_dataset['ID']):
  if item[:5]=='train':
    is_train.append(True)
    is_dev.append(False)
  else:
    is_train.append(False)
    is_dev.append(True)

train_q= q_dataset[is_train]
dev_q= q_dataset[is_dev]
```

```python
#ID 숫자로 바꿔주기

def df_to_sortednum(dev_q):
  int_dev_q= dev_q.copy()
  int_dev_q= int_dev_q.reset_index()
  int_dev_q.drop(['index'], axis=1, inplace= True)

  if int_dev_q['ID'][0][:3]=='dev':
    split= 4
  elif int_dev_q['ID'][0][:5]=='train':
    split= 6

  for index, item in enumerate(int_dev_q['ID']):
    number= int(item[split:])
    int_dev_q['ID'][index]= number


#ID 오름차순으로 정리

  int_dev_q= int_dev_q.sort_values(by= 'ID',ascending=True)
  int_dev_q= int_dev_q.reset_index()
  int_dev_q.drop(['index'], axis=1, inplace= True)
  
  return int_dev_q
```

```python
int_dev_q= df_to_sortednum(dev_q)
int_train_q= df_to_sortednum(train_q)
int_dev_text= df_to_sortednum(dev_text)
int_train_text= df_to_sortednum(train_text)

```

```python
int_train_text
```

```python

train_data['Text NER']= int_train_text['NER']
train_data['Question NER']= int_train_q['NER']
train_data.to_csv('SKT_BoolQ_NER_Train.tsv', index= False, header= True, sep= '\t')

```

```python

dev_data['Text NER']= int_dev_text['NER']
dev_data['Question NER']= int_dev_q['NER']
dev_data.to_csv('SKT_BoolQ_NER_Dev.tsv', index= False, header= True, sep= '\t')
```
