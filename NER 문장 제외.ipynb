```python
df= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/국립국어원/SKT_BoolQ_NER_Dev.tsv', delimiter= '\t')

def drop_ner(df):
  proper_text=[[] for _ in range(len(df['Text']))]
  count=0

  for index, text in enumerate(df['Text']):
    for t in text.split('.')[:-1]:
      proper_text[index].append(t + '.')

    
    if pd.isnull(df['Question NER'][index]):
      pass
  
    else:
      if len(proper_text[index])>=4:
        count+=1
        tokens= df['Question NER'][index].split(' ')
        for sentence in proper_text[index]:
          flag = 0
          for word in tokens:
            if word in sentence:
              flag = 1
          if flag==0:
            proper_text[index].remove(sentence)
        
        

  count2=0
  for item in proper_text:
    if len(item)>=4:
      count2+=1
  print('before :', count, 'after :', count2)

  return proper_text

def merge_and_save(df, textlist):

  for index, text_list in enumerate(textlist):
    text= ''
    for texts in text_list:
      text+=texts

    textlist[index]= text

  df['Text']= textlist
  print(df.head())
  df.to_csv('SKT_NER_DROP_Dev.tsv', index=False, header= True, sep='\t')


textlist= drop_ner(df)
merge_and_save(df,textlist)
```
